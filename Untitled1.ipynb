```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
m_tau = 1.776  # Tau mass in GeV
m_pi = 0.139   # Charged pion mass in GeV
m_Z = 91.1876  # Z-boson mass in GeV
sigma_energy = 0.02  # Energy resolution for Gaussian uncertainty (20%)
sigma_angle = 0.01   # Angular resolution in radians

# Generate truth data for tau decays
def generate_tau_event():
    # Generate random 4-momentum for the Z boson (rest frame, so Z is at rest)
    p_Z = np.array([m_Z, 0, 0, 0])  # [E, px, py, pz]

    # Generate random angles for tau decay directions
    theta_tau1 = np.random.uniform(0, np.pi)
    phi_tau1 = np.random.uniform(0, 2 * np.pi)
    
    theta_tau2 = np.pi - theta_tau1  # Back-to-back in the Z rest frame
    phi_tau2 = (phi_tau1 + np.pi) % (2 * np.pi)

    # Tau energies (equal in the Z rest frame)
    E_tau = m_Z / 2
    p_tau_mag = np.sqrt(E_tau**2 - m_tau**2)
    
    # Tau 4-momenta
    p_tau1 = np.array([
        E_tau,
        p_tau_mag * np.sin(theta_tau1) * np.cos(phi_tau1),
        p_tau_mag * np.sin(theta_tau1) * np.sin(phi_tau1),
        p_tau_mag * np.cos(theta_tau1)
    ])
    p_tau2 = np.array([
        E_tau,
        p_tau_mag * np.sin(theta_tau2) * np.cos(phi_tau2),
        p_tau_mag * np.sin(theta_tau2) * np.sin(phi_tau2),
        p_tau_mag * np.cos(theta_tau2)
    ])

    # Generate neutrino and pion momenta in tau rest frames
    E_pi = (m_tau**2 + m_pi**2) / (2 * m_tau)
    p_pi_mag = np.sqrt(E_pi**2 - m_pi**2)
    
    # Randomize pion directions in tau1 rest frame
    theta_pi1 = np.random.uniform(0, np.pi)
    phi_pi1 = np.random.uniform(0, 2 * np.pi)
    p_pi1_tau1 = np.array([
        E_pi,
        p_pi_mag * np.sin(theta_pi1) * np.cos(phi_pi1),
        p_pi_mag * np.sin(theta_pi1) * np.sin(phi_pi1),
        p_pi_mag * np.cos(theta_pi1)
    ])
    
    # Back-to-back neutrino in tau1 rest frame
    p_nu1_tau1 = np.array([m_tau - E_pi, -p_pi1_tau1[1], -p_pi1_tau1[2], -p_pi1_tau1[3]])

    # Same for tau2
    theta_pi2 = np.random.uniform(0, np.pi)
    phi_pi2 = np.random.uniform(0, 2 * np.pi)
    p_pi2_tau2 = np.array([
        E_pi,
        p_pi_mag * np.sin(theta_pi2) * np.cos(phi_pi2),
        p_pi_mag * np.sin(theta_pi2) * np.sin(phi_pi2),
        p_pi_mag * np.cos(theta_pi2)
    ])
    p_nu2_tau2 = np.array([m_tau - E_pi, -p_pi2_tau2[1], -p_pi2_tau2[2], -p_pi2_tau2[3]])

    # Boost pion and neutrino momenta to the lab frame (Z rest frame)
    def boost_to_lab(p_tau, p_particle):
        gamma = p_tau[0] / m_tau
        beta = np.linalg.norm(p_tau[1:]) / p_tau[0]
        beta_vec = p_tau[1:] / np.linalg.norm(p_tau[1:])
        
        p_par = np.dot(p_particle[1:], beta_vec)
        p_perp = p_particle[1:] - p_par * beta_vec
        E_lab = gamma * (p_particle[0] - beta * p_par)
        p_par_lab = gamma * (p_par - beta * p_particle[0])
        p_lab = np.concatenate(([E_lab], p_perp + p_par_lab * beta_vec))
        return p_lab

    p_pi1 = boost_to_lab(p_tau1, p_pi1_tau1)
    p_nu1 = boost_to_lab(p_tau1, p_nu1_tau1)
    p_pi2 = boost_to_lab(p_tau2, p_pi2_tau2)
    p_nu2 = boost_to_lab(p_tau2, p_nu2_tau2)

    return p_tau1, p_tau2, p_pi1, p_pi2, p_nu1, p_nu2

# Generate reconstructed data with Gaussian uncertainties
def add_uncertainties(p):
    E = p[0]
    px, py, pz = p[1:]
    E_reco = np.random.normal(E, sigma_energy * E)
    theta = np.arctan2(np.sqrt(px**2 + py**2), pz)
    phi = np.arctan2(py, px)
    theta_reco = np.random.normal(theta, sigma_angle)
    phi_reco = np.random.normal(phi, sigma_angle)
    p_mag = np.sqrt(E_reco**2 - m_pi**2)
    px_reco = p_mag * np.sin(theta_reco) * np.cos(phi_reco)
    py_reco = p_mag * np.sin(theta_reco) * np.sin(phi_reco)
    pz_reco = p_mag * np.cos(theta_reco)
    return np.array([E_reco, px_reco, py_reco, pz_reco])

# Simulate events
n_events = 500
truth_data = []
reco_data = []
for _ in range(n_events):
    p_tau1, p_tau2, p_pi1, p_pi2, p_nu1, p_nu2 = generate_tau_event()
    truth_data.append((p_tau1, p_tau2, p_pi1, p_pi2, p_nu1, p_nu2))
    reco_data.append((add_uncertainties(p_pi1), add_uncertainties(p_pi2)))

# Plot results
true_energies = [p[2][0] for p in truth_data] + [p[3][0] for p in truth_data]
reco_energies = [p[0][0] for p in reco_data] + [p[1][0] for p in reco_data]

plt.hist(true_energies, bins=50, alpha=0.7, label="True Pion Energies")
plt.hist(reco_energies, bins=50, alpha=0.7, label="Reco Pion Energies")
plt.xlabel("Energy (GeV)")
plt.ylabel("Count")
plt.legend()
plt.title("True vs. Reconstructed Pion Energies")
plt.show()
```

```python
# Extract neutrino data for plotting
true_neutrino_energies = [p[4][0] for p in truth_data] + [p[5][0] for p in truth_data]
true_neutrino_theta = []
true_neutrino_phi = []

# Calculate theta and phi for neutrinos
for event in truth_data:
    for i in [4, 5]:  # Neutrinos
        px, py, pz = event[i][1:]
        theta = np.arctan2(np.sqrt(px**2 + py**2), pz)
        phi = np.arctan2(py, px)
        true_neutrino_theta.append(theta)
        true_neutrino_phi.append(phi)

# Plot neutrino energy distribution
plt.figure(figsize=(10, 6))
plt.hist(true_neutrino_energies, bins=50, alpha=0.7, color='blue', label='Neutrino Energies')
plt.xlabel('Energy (GeV)')
plt.ylabel('Count')
plt.title('Neutrino Energy Distribution')
plt.legend()
plt.show()

# Plot neutrino theta distribution
plt.figure(figsize=(10, 6))
plt.hist(true_neutrino_theta, bins=50, alpha=0.7, color='blue', label='Neutrino Theta')
plt.xlabel(r'$\theta$ (radians)')
plt.ylabel('Count')
plt.title('Neutrino Theta Distribution')
plt.legend()
plt.show()

# Plot neutrino phi distribution
plt.figure(figsize=(10, 6))
plt.hist(true_neutrino_phi, bins=50, alpha=0.7, color='blue', label='Neutrino Phi')
plt.xlabel(r'$\phi$ (radians)')
plt.ylabel('Count')
plt.title('Neutrino Phi Distribution')
plt.legend()
plt.show()
```

```python
# Function to normalize angles to the range [-pi, pi]
def normalize_angle(angle):
    return (angle + np.pi) % (2 * np.pi) - np.pi

# Function to calculate theta and phi from a 4-momentum
def calculate_angles(p):
    px, py, pz = p[1], p[2], p[3]
    theta = np.arctan2(np.sqrt(px**2 + py**2), pz)
    phi = np.arctan2(py, px)
    return theta, phi

# Reconstruct tau momenta from reconstructed pion and truth neutrino data
def reconstruct_tau(p_pi, p_nu):
    return p_pi + p_nu  # Sum the pion and neutrino 4-momenta

# Collect residuals for tau angles
theta_residuals = []
phi_residuals = []

for truth, reco in zip(truth_data, reco_data):
    p_tau1, p_tau2, p_pi1, p_pi2, p_nu1, p_nu2 = truth
    p_pi1_reco, p_pi2_reco = reco

    # Reconstruct taus
    p_tau1_reco = reconstruct_tau(p_pi1_reco, p_nu1)
    p_tau2_reco = reconstruct_tau(p_pi2_reco, p_nu2)

    # Calculate angles for truth taus
    theta_tau1, phi_tau1 = calculate_angles(p_tau1)
    theta_tau2, phi_tau2 = calculate_angles(p_tau2)

    # Calculate angles for reconstructed taus
    theta_tau1_reco, phi_tau1_reco = calculate_angles(p_tau1_reco)
    theta_tau2_reco, phi_tau2_reco = calculate_angles(p_tau2_reco)

    # Calculate residuals
    theta_residuals.extend([theta_tau1 - theta_tau1_reco, theta_tau2 - theta_tau2_reco])
    phi_residuals.extend([
        normalize_angle(phi_tau1 - phi_tau1_reco),
        normalize_angle(phi_tau2 - phi_tau2_reco)
    ])

# Plot theta residuals
plt.figure(figsize=(12, 6))
plt.hist(theta_residuals, bins=50, alpha=0.7, label=r"$\Delta \theta$ (Tau)", color="blue")
plt.xlabel(r"$\Delta \theta$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Tau Theta Angles")
plt.legend()
plt.show()

# Plot phi residuals
plt.figure(figsize=(12, 6))
plt.hist(phi_residuals, bins=50, alpha=0.7, label=r"$\Delta \phi$ (Tau)", color="orange")
plt.xlabel(r"$\Delta \phi$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Tau Phi Angles")
plt.legend()
plt.show()
```

```python
from scipy.optimize import minimize
import numpy as np
from tqdm import tqdm

# Chi-squared function
def chi_squared(params, p_pi1, p_pi2):
    # Extract the neutrino parameters
    px_nu1, py_nu1, pz_nu1 = params[0], params[1], params[2]
    px_nu2, py_nu2, pz_nu2 = params[3], params[4], params[5]

    # Reconstruct the neutrino 4-momenta
    E_nu1 = np.sqrt(px_nu1**2 + py_nu1**2 + pz_nu1**2)  # Assuming massless neutrinos
    E_nu2 = np.sqrt(px_nu2**2 + py_nu2**2 + pz_nu2**2)

    p_nu1 = np.array([E_nu1, px_nu1, py_nu1, pz_nu1])
    p_nu2 = np.array([E_nu2, px_nu2, py_nu2, pz_nu2])

    # Reconstruct the tau 4-momenta
    p_tau1 = p_pi1 + p_nu1
    p_tau2 = p_pi2 + p_nu2

    # Tau mass constraint
    tau1_mass_constraint = (p_tau1[0]**2 - np.sum(p_tau1[1:]**2) - m_tau**2)**2
    tau2_mass_constraint = (p_tau2[0]**2 - np.sum(p_tau2[1:]**2) - m_tau**2)**2

    # Transverse momentum conservation
    px_total = px_nu1 + px_nu2 + p_pi1[1] + p_pi2[1]
    py_total = py_nu1 + py_nu2 + p_pi1[2] + p_pi2[2]

    pT_constraint = px_total**2 + py_total**2

    # Invariant mass of the tau-tau system
    p_Z = p_tau1 + p_tau2
    invariant_mass_constraint = (p_Z[0]**2 - np.sum(p_Z[1:]**2) - m_Z**2)**2

    # Total chi-squared
    chi2 = (
        tau1_mass_constraint / sigma_energy**2
        + tau2_mass_constraint / sigma_energy**2
        + pT_constraint / sigma_energy**2
        + invariant_mass_constraint / sigma_energy**2
    )
    return chi2

# Perform chi-squared fitting for all events
fitted_nu_momenta = []
fitted_tau_momenta = []

for truth, reco in tqdm(zip(truth_data, reco_data)):
    p_pi1_reco, p_pi2_reco = reco  # Reconstructed pion momenta

    # Initial guess for neutrino momenta
    initial_guess = np.zeros(6)

    # Minimize chi-squared
    result = minimize(
        chi_squared,
        initial_guess,
        args=(p_pi1_reco, p_pi2_reco),
        method="L-BFGS-B"
    )

    # Extract fitted neutrino momenta
    fitted_params = result.x
    px_nu1, py_nu1, pz_nu1 = fitted_params[0], fitted_params[1], fitted_params[2]
    px_nu2, py_nu2, pz_nu2 = fitted_params[3], fitted_params[4], fitted_params[5]

    E_nu1 = np.sqrt(px_nu1**2 + py_nu1**2 + pz_nu1**2)
    E_nu2 = np.sqrt(px_nu2**2 + py_nu2**2 + pz_nu2**2)

    p_nu1_fit = np.array([E_nu1, px_nu1, py_nu1, pz_nu1])
    p_nu2_fit = np.array([E_nu2, px_nu2, py_nu2, pz_nu2])

    # Reconstruct tau momenta
    p_tau1_fit = p_pi1_reco + p_nu1_fit
    p_tau2_fit = p_pi2_reco + p_nu2_fit

    # Append results
    fitted_nu_momenta.append((p_nu1_fit, p_nu2_fit))
    fitted_tau_momenta.append((p_tau1_fit, p_tau2_fit))
```

```python
# Initialize lists for storing residuals
theta_residuals = []
phi_residuals = []

for truth, reco, fit in zip(truth_data, reco_data, fitted_nu_momenta):
    # Extract true and reconstructed neutrino momenta
    p_nu1_truth, p_nu2_truth = truth[4], truth[5]
    p_nu1_fit, p_nu2_fit = fit

    # Calculate true angles for neutrinos
    theta_nu1_truth, phi_nu1_truth = calculate_angles(p_nu1_truth)
    theta_nu2_truth, phi_nu2_truth = calculate_angles(p_nu2_truth)

    # Calculate fitted angles for neutrinos
    theta_nu1_fit, phi_nu1_fit = calculate_angles(p_nu1_fit)
    theta_nu2_fit, phi_nu2_fit = calculate_angles(p_nu2_fit)

    # Compute residuals (truth - fit)
    theta_residuals.extend([theta_nu1_truth - theta_nu1_fit, theta_nu2_truth - theta_nu2_fit])
    phi_residuals.extend([phi_nu1_truth - phi_nu1_fit, phi_nu2_truth - phi_nu2_fit])

# Plot theta residuals
plt.figure(figsize=(12, 6))
plt.hist(theta_residuals, bins=50, alpha=0.7, label=r"$\Delta \theta$ (Neutrino)", color="blue")
plt.xlabel(r"$\Delta \theta$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Neutrino Theta Angles")
plt.legend()
plt.show()

# Plot phi residuals
plt.figure(figsize=(12, 6))
plt.hist(phi_residuals, bins=50, alpha=0.7, label=r"$\Delta \phi$ (Neutrino)", color="orange")
plt.xlabel(r"$\Delta \phi$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Neutrino Phi Angles")
plt.legend()
plt.show()
```

```python
# Function to normalize angles to the range [-pi, pi]
def normalize_angle(angle):
    return (angle + np.pi) % (2 * np.pi) - np.pi

# Initialize lists for storing residuals
theta_tau_residuals = []
phi_tau_residuals = []

for truth, fit in zip(truth_data, fitted_tau_momenta):
    # Extract true and reconstructed tau momenta
    p_tau1_truth, p_tau2_truth = truth[0], truth[1]
    p_tau1_fit, p_tau2_fit = fit

    # Calculate true angles for taus
    theta_tau1_truth, phi_tau1_truth = calculate_angles(p_tau1_truth)
    theta_tau2_truth, phi_tau2_truth = calculate_angles(p_tau2_truth)

    # Calculate fitted angles for taus
    theta_tau1_fit, phi_tau1_fit = calculate_angles(p_tau1_fit)
    theta_tau2_fit, phi_tau2_fit = calculate_angles(p_tau2_fit)

    # Compute residuals (truth - fit) and normalize phi residuals to [-pi, pi]
    theta_tau_residuals.extend([theta_tau1_truth - theta_tau1_fit, theta_tau2_truth - theta_tau2_fit])
    phi_tau_residuals.extend([
        normalize_angle(phi_tau1_truth - phi_tau1_fit),
        normalize_angle(phi_tau2_truth - phi_tau2_fit)
    ])

# Plot theta residuals
plt.figure(figsize=(12, 6))
plt.hist(theta_tau_residuals, bins=50, alpha=0.7, label=r"$\Delta \theta$ (Tau)", color="blue")
plt.xlabel(r"$\Delta \theta$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Tau Theta Angles")
plt.legend()
plt.show()

# Plot phi residuals
plt.figure(figsize=(12, 6))
plt.hist(phi_tau_residuals, bins=50, alpha=0.7, label=r"$\Delta \phi$ (Tau)", color="orange")
plt.xlabel(r"$\Delta \phi$ (radians)")
plt.ylabel("Count")
plt.title("Residuals for Tau Phi Angles")
plt.legend()
plt.show()
```
